```python
x=[0,1,2,3,4]
y=[i**2 for i in x]
plt.plot(x,y)
fig,ax=plt.subplots()
ax.plot(x,y,marker='o',label="Data points")
ax.set_title("Basic component of matplot fig")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
plt.show()
```

```python
x=[0,1,2,3,4]
y=[i**2 for i in x]
y1=[i**2 for i in x]
y2=[i**3 for i in x]
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.plot(x,y1,marker=">",label="data points")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.plot(x,y2,marker=">",label="data points")
plt.show()
```
